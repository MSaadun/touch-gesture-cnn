```python
import numpy
import matplotlib.pyplot as plt
from pyts.image import GramianAngularField
from pathlib import Path
```

# Variant 1

```python
take = 0
eerste = 1
a = numpy.empty((0, 68))
i = 0
c = 0

gasf = GramianAngularField(method='summation')
entries = Path('Full/gesture14/variant1/')

for entry in entries.iterdir():
    df2 = numpy.loadtxt(open(entry, "rb"), delimiter=",", dtype = int, skiprows=1)
    last_row = df2[-1]
    take = 0 
    c = 0
    d = None
    dfT = None
    X_gasf = None
    for row in df2:
        if row[3] == 1 and eerste == 1 and i==0 and take == 0:
            a = numpy.vstack([a, row])
            eerste += 1
        elif eerste >1 and row[3] != 1 and i==0:
            a = numpy.vstack([a, row])
        elif row[3] == 1 and eerste >1 and i==0:
            fig, ax = plt.subplots(nrows = 8, ncols = 8, figsize=(5, 5), dpi = 50)
            fig.subplots_adjust(hspace=0.0, wspace=0.0)
            for x in ax.ravel():
                x.axis("off")
            take += 1
            d = None
            dfT = None
            X_gasf = None
            d = a[:,4:]
            dfT = d.transpose()
            X_gasf = gasf.fit_transform(dfT)
            for t in range(8):
                for j in range(8):
                    ax[t, j].imshow(X_gasf[c], cmap='rainbow', origin='lower' )
                    c += 1
                fig.savefig('Images/gaf/gesture14/variant1/{}_{}_{}_{}.png'.format(row[0],row[1],row[2], take))
            a = numpy.empty((0, 68))
            i += 1
            a = numpy.vstack([a, row])
            eerste = 1
            plt.close('all')
        elif row[3] != 1 and i==1:
            a = numpy.vstack([a, row])
        elif row[3] == 1 and i==1:
            fig, ax = plt.subplots(nrows = 8, ncols = 8, figsize=(5, 5), dpi = 50)
            fig.subplots_adjust(hspace=0.0, wspace=0.0)
            for x in ax.ravel():
                x.axis("off")
            take += 1
            d = None
            dfT = None
            X_gasf = None
            c = 0
            d = a[:,4:]
            dfT = d.transpose()
            X_gasf = gasf.fit_transform(dfT)
            for t in range(8):
                for j in range(8):
                    ax[t, j].imshow(X_gasf[c], cmap='rainbow', origin='lower' )
                    c += 1
                fig.savefig('Images/gaf/gesture14/variant1/{}_{}_{}_{}.png'.format(row[0],row[1],row[2], take))
            a = numpy.empty((0, 68))
            i += 1
            a = numpy.vstack([a, row])
            plt.close('all')
        elif row[3] != 1 and i==2:
            a = numpy.vstack([a, row])
        elif row[3] == 1 and i==2:
            fig, ax = plt.subplots(nrows = 8, ncols = 8, figsize=(5, 5), dpi = 50)
            fig.subplots_adjust(hspace=0.0, wspace=0.0)
            for x in ax.ravel():
                x.axis("off")
            take += 1
            d = None
            dfT = None
            X_gasf = None
            c = 0
            d = a[:,4:]
            dfT = d.transpose()
            X_gasf = gasf.fit_transform(dfT)
            for t in range(8):
                for j in range(8):
                    ax[t, j].imshow(X_gasf[c], cmap='rainbow', origin='lower' )
                    c += 1
                fig.savefig('Images/gaf/gesture14/variant1/{}_{}_{}_{}.png'.format(row[0],row[1],row[2], take))
            a = numpy.empty((0, 68))
            i += 1
            a = numpy.vstack([a, row])
            plt.close('all')
        elif row[3] != 1 and i==3:
            a = numpy.vstack([a, row])
        elif row[3] == 1 and i==3:
            fig, ax = plt.subplots(nrows = 8, ncols = 8, figsize=(5, 5), dpi = 50)
            fig.subplots_adjust(hspace=0.0, wspace=0.0)
            for x in ax.ravel():
                x.axis("off")
            take += 1
            d = None
            dfT = None
            X_gasf = None
            c = 0
            d = a[:,4:]
            dfT = d.transpose()
            X_gasf = gasf.fit_transform(dfT)
            for t in range(8):
                for j in range(8):
                    ax[t, j].imshow(X_gasf[c], cmap='rainbow', origin='lower' )
                    c += 1
                fig.savefig('Images/gaf/gesture14/variant1/{}_{}_{}_{}.png'.format(row[0],row[1],row[2], take))
            a = numpy.empty((0, 68))
            i += 1
            a = numpy.vstack([a, row])
            plt.close('all')
        elif row[3] != 1 and i==4:
            a = numpy.vstack([a, row])
        elif row[3] == 1 and i==4:
            fig, ax = plt.subplots(nrows = 8, ncols = 8, figsize=(5, 5), dpi = 50)
            fig.subplots_adjust(hspace=0.0, wspace=0.0)
            for x in ax.ravel():
                x.axis("off")
            take += 1
            d = None
            dfT = None
            X_gasf = None
            c = 0
            d = a[:,4:]
            dfT = d.transpose()
            X_gasf = gasf.fit_transform(dfT)
            for t in range(8):
                for j in range(8):
                    ax[t, j].imshow(X_gasf[c], cmap='rainbow', origin='lower' )
                    c += 1
                fig.savefig('Images/gaf/gesture14/variant1/{}_{}_{}_{}.png'.format(row[0],row[1],row[2], take))
            a = numpy.empty((0, 68))
            i += 1 
            a = numpy.vstack([a, row])
            plt.close('all')
            take += 1
            laatste = 1
        elif row[3] != 1 and i==5 and take == 6 and laatste == 1:
            if numpy.array_equal(row,last_row) == False:
                a = numpy.vstack([a, row])
            if numpy.array_equal(row,last_row):
                a = numpy.vstack([a, row])
                i = 0
                fig, ax = plt.subplots(nrows = 8, ncols = 8, figsize=(5, 5), dpi = 50)
                fig.subplots_adjust(hspace=0.0, wspace=0.0)
                for x in ax.ravel():
                    x.axis("off")
                d = None
                dfT = None
                X_gasf = None
                c = 0
                d = a[:,4:]
                dfT = d.transpose()
                X_gasf = gasf.fit_transform(dfT)
                for t in range(8):
                    for j in range(8):
                        ax[t, j].imshow(X_gasf[c], cmap='rainbow', origin='lower' )
                        c += 1
                    fig.savefig('Images/gaf/gesture14/variant1/{}_{}_{}_{}.png'.format(row[0],row[1],row[2], take))
                take = 0
                laatste = 0
                plt.close('all')
                a = numpy.empty((0, 68))
                break

```
